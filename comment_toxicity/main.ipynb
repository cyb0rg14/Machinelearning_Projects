## Converting `text to vectors` to work as `tensors` for model

```python
!pip install sentence_transformers
```

```python
from sentence_transformers import SentenceTransformer

model0 = SentenceTransformer('sentence-transformers/all-MiniLM-L6-v2')
```

```python
embeddings = model0.encode(train['comment_text'])
```

```python
embeddings
```

```python
embeddings.shape
```

```python
embeddings = embeddings.reshape(159571, 1, 384)
embeddings.shape
```

```python
Y_train = train.iloc[:, 2:].values
```

```python
Y_train.shape
```

## Now time to create the model using `keras`

```python
import tensorflow as tf
from tensorflow import keras
from keras import Sequential
from keras.layers import Dense, Dropout, Bidirectional, LSTM
```

```python
model = Sequential()

model.add(Bidirectional(LSTM(units=64, activation='tanh', input_shape=(1, 384))))
model.add(Dense(128, activation='relu'))
model.add(Dense(256, activation='relu'))
model.add(Dense(128, activation='relu'))
model.add(Dense(6, activation='sigmoid'))
```

```python
model.compile(optimizer="adam", loss="BinaryCrossentropy", metrics=["accuracy"])
```

```python
model.summary()
```

```python
history = model.fit(embeddings, Y_train, epochs=5, validation_split=0.2)
```

```python
test_embeddings = model0.encode(test.iloc[0].comment_text)
```

```python
test_embeddings = test_embeddings.reshape(1, 1, 384)
```

```python
y_pred = model.predict(test_embeddings)
```

```python
def score_comment(comment: str):
    embeddings = model0.encode(comment)
    embeddings = embeddings.reshape(1, 1, 384)
    results = model.predict(embeddings)

    text = ''
    for idx, col in enumerate(test_labels.columns[1:]):
        text += f"{col}: {results[0][idx]>0.5}\n"

    return text
```
